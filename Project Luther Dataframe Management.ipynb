```python
import pandas as pd
import os
```

```python
#initializes working df
master_df = pd.read_pickle('P2pickles/1-2003df.pkl')
master_df['year'] = 2003
master_df.tail()
```

```python
#finding files
files = os.listdir('P2pickles')
pickles = []
for file in files:
    if 'pkl' in file:
        pickles.append(file)
```

```python
#Reads in all pickled dataframe files from scraper

path = 'P2pickles/'
for file in pickles:
    #print(path+file)
    df = pd.read_pickle(path+file)
    df['year'] = int(file[2:6])
    #print(df.describe())
    master_df = master_df.append(df, sort=False)
master_df.head()
```

## Cleaning and Feature Engineering on new Master Dataframe
Here, we will be removing all of the faulty data that was collected by the scraper and normalizing the rest. Normalization has to occur so that the trend between playing time and statistics can be seen.  If a player has a low playing time, it makes sense that they will have low statistics otherwise, so we will draw inference from the rates of their statistics rather than the raw values. 

```python
df = master_df
#Clearing off poor values
df = df[df.fg!=-1]
df.sample(20)
```

```python

#Converting Statistics into Rates
df['fg_per_mp'] = df.fg/df.minutes_played
df['fga_per_mp'] = df.fga/df.minutes_played
df['fg3_per_mp'] = df.fg3/df.minutes_played
df['fg3a_per_mp'] = df.fg3a/df.minutes_played
df['ft_per_mp'] = df.ft/df.minutes_played
df['fta_per_mp'] = df.fta/df.minutes_played
df['or_per_mp'] = df['or']/df.minutes_played
df['dr_per_mp'] = df.dr/df.minutes_played
df['tot_r_per_mp'] = df.tot_r/df.minutes_played
df['asst_per_mp'] = df.asst/df.minutes_played
df['steals_per_mp'] = df.steals/df.minutes_played
df['blocks_per_mp'] = df['blocks']/df.minutes_played
df['turnovers_per_mp'] = df.turnovers/df.minutes_played
df['fouls_per_mp'] = df.fouls/df.minutes_played
df['points_per_mp'] = df.points/df.minutes_played
df['2s_per_mp']=df.fg_per_mp-df.fg3_per_mp
df['fg2_pct'] = (df.fg-df.fg3)/(df.fga-df.fg3a)

df.head()
```

```python
df = df[['minutes_played', 'fg_pct', '3_pct', 'ft_pct', 'fg_per_mp', 'fga_per_mp', 'fg3_per_mp', 'fg3a_per_mp', 
  'ft_per_mp', 'fta_per_mp', 'or_per_mp', 'dr_per_mp', 'tot_r_per_mp', 'asst_per_mp', 'steals_per_mp',
  'blocks_per_mp', 'turnovers_per_mp', 'fouls_per_mp', 'points_per_mp', '2s_per_mp','year', 'fg2_pct']]
```

## Cells after this are for after the initial dataframe has been constructed
The following are for reading and operating on 'master_df.pkl'

```python
pt_df = pd.read_pickle('master_df.pkl')
pt_df['name'] = pt_df.index
pt_df.head()
```

```python
cash_df = pd.DataFrame.from_csv('nba_salaries_1990_to_2018.csv')
cleaned_names = []
for name in cash_df.index:
    try:
        new_name = name.split()[1] +','+name.split()[0]
    except:
        new_name = name
    cleaned_names.append(new_name)
cash_df['name'] = cleaned_names
cash_df.head()
```

```python
cash_df = cash_df[['salary', 'season_end', 'team', 'name']]
cash_df.head()
```

```python
df = pd.merge(pt_df,cash_df, how='left', left_on=['name','year'], right_on=['name','season_end'])
df.describe()
```

```python
df = df.dropna()
df.head()
```

```python
df.to_pickle('merged_df.pkl')
```
